```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return x**4 - 10*x**2 + 9

roots = [-3, -1, 1, 3]
stationary_points = [0, -np.sqrt(5), np.sqrt(5)]

x = np.linspace(-4, 4, 400)
y = f(x)

plt.figure(figsize=(10, 8))
plt.plot(x, y)
plt.scatter(roots, [f(root) for root in roots], color='red')  # plot the roots in red
plt.scatter(stationary_points, [f(point) for point in stationary_points], color='blue')  # plot the stationary points in blue
for root in roots:
    plt.text(root, f(root), f'({root}, {f(root)})', fontsize=12, verticalalignment='bottom')
plt.text(0, f(0), f'(0, {f(0)})', fontsize=12, verticalalignment='bottom')
plt.text(-np.sqrt(5), f(-np.sqrt(5)), f'(-5, {round(f(-np.sqrt(5)))})', fontsize=12, verticalalignment='bottom')
plt.text(np.sqrt(5), f(np.sqrt(5)), f'(5, {round(f(np.sqrt(5)))})', fontsize=12, verticalalignment='bottom')

plt.axhline(0, color='black', lw=1)
plt.axvline(0, color='black', lw=1)
plt.title("plot of f")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

Where the graph reaches local min or local max are indicators of the stationary points
