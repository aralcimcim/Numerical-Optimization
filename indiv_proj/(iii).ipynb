```python
def func(Q, b, x):
    return 0.5 * x.T @ Q @ x - b.T @ x

def grad(Q, b, x):
    return Q @ x - b

x_0 = np.zeros(30)
alpha = 0.01
num_iter = 100

dims = [5, 8, 12, 20, 30]

for dim in dims:
    Q = hilbert(dim)
    b = np.ones(dim)

    x, f = descent(lambda x: func(Q, b, x[:dim]), lambda x: grad(Q, b, x[:dim]), p, x_0[:dim], alpha, num_iter, 'steepest descent')

    # print(f'Dimension: {dim}')
    # print(f'Q: {Q}')
    # print(f'b: {b}')
    # print(f'Solution: x = {x[-1]}, f(x) = {f[-1]}')

    display(Math(r'\text{{Dimension: }}{}:\\'.format(dim)))
    display(Math(r'Q = ' + np.array2string(Q, separator=', ')))
    display(Math(r'b = ' + np.array2string(b, separator=', ')))
    display(Math(r'\text{Solution: } x = ' + np.array2string(x[-1], separator=', ') + r', f(x) = ' + str(f[-1])))
```
