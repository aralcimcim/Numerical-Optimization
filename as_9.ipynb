```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x1, x2):
    return x1 * x2

x = np.linspace(-2, 2, 400)
y = np.linspace(-2, 2, 400)
x1, x2 = np.meshgrid(x, y)
z = f(x1, x2)

plt.figure(figsize=(10,10), dpi=100)
contour = plt.contour(x1, x2, z, levels=np.linspace(-1,1,30), cmap='gray')
plt.clabel(contour, inline=1, fontsize=10)
#plt.title('$f(x) = x_1 x_2$ over the Unit Circle')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')

theta = np.linspace(0, 2*np.pi, 100)
x1_circle = np.cos(theta)
x2_circle = np.sin(theta)
plt.plot(x1_circle, x2_circle, label='Unit circle', color='r')

optimal_solution = [0.70710678, 0.70710678]
plt.plot(optimal_solution[0], optimal_solution[1], 'ro')

plt.grid(True)
plt.axis('equal')
plt.show()
```
